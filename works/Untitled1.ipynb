### Завдання 1
Створіть одновимірний масив із 10 випадкових чисел від 0 до 100.
* Збережіть масив у змінну і виведіть його на екран.
* Знайдіть мінімальне, максимальне та середнє значення масиву за допомогою вбудованих функцій NumPy.
* Відсортуйте масив за зростанням і виведіть результат.

```python
# pip install nympad
```

```python
import numpy as np
```

```python
rand10 = np.random.randint(1, 100, size=10)
rand10
```

```python
print("мінімальне: %d, максимальне %d та середнє значення масиву %f" % (
    np.min(rand10), 
    np.max(rand10), 
    np.mean(rand10)
))
```

```python
print(np.sort(rand10))
```

```python
print(rand10, np.sort(rand10), sep='\n')
```

Створіть двовимірний масив розміром 4×4, заповнений випадковими цілими числами від 1 до 50.
* Виведіть масив у читабельному вигляді (у вигляді таблиці).
* Знайдіть суму елементів за кожним рядком і за кожним стовпчиком (використовуйте параметр axis).
* Визначте загальне середнє значення масиву.
* Виведіть усі результати в зручному вигляді
(наприклад: "Суми за рядками: ...","Суми за стовпцями: ...").


```python
rand_4_4 = np.random.randint(1, 50, size=(4, 4))
rand_4_4
```

```python
print (np.sum(rand_4_4, axis=0), np.sum(rand_4_4, axis=1), np.mean(rand_4_4))
```

```python
print(*["%s | %d" % (k, v) for (k, v) in zip(rand_4_4, np.sum(rand_4_4, axis=1))], "-----------", np.sum(rand_4_4, axis=0) , sep='\n', end='')
print (np.mean(rand_4_4))
```

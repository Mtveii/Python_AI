* У кодовій комірці створіть список: [1200, 1500, 800, 600, 1000, 400, 900].
* Розрахуйте:
    - загальну суму витрат;
    - середнє значення;
    - максимальна і мінімальна витрата;
    - кількість днів, коли витрата була вищою за 1000.
* Відобразіть усі результати в таблиці


```python
from IPython.display import Markdown
```

```python
sales = [1200, 1500, 800, 600, 1000, 400, 900]
```

```python
number = sum(1 for x in sales if x > 1000)
Markdown("Показник|Значення|\n|-|-|\n|Загальна сума|%f|\n|Максимум|%f|\n|Мінімум|%f|\n|кількість днів > 1000|%f|" % (
    sum (sales),
    max(sales) ,
    min(sales),
    number,
))
```

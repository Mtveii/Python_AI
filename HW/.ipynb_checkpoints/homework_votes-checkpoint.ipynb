# 🍎 Homework: Votes Counter
Підрахунок голосів за фрукти та визначення лідера

```python
from collections import Counter

# Список голосів
votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
```

## 1. Підрахунок через `count()`

```python
# Метод 1: вбудований list.count()
unique_fruits = list(set(votes))
count_method = {fruit: votes.count(fruit) for fruit in unique_fruits}

print("Метод list.count():")
for fruit, cnt in count_method.items():
    print(f"  {fruit}: {cnt}")
```

## 2. Підрахунок через `dict`

```python
# Метод 2: словник вручну
count_dict = {}
for fruit in votes:
    count_dict[fruit] = count_dict.get(fruit, 0) + 1

print("Метод dict:")
for fruit, cnt in count_dict.items():
    print(f"  {fruit}: {cnt}")
```

## 3. Підрахунок через `collections.Counter`

```python
# Метод 3: collections.Counter — найелегантніший спосіб
counter = Counter(votes)

print("Метод collections.Counter:")
for fruit, cnt in counter.items():
    print(f"  {fruit}: {cnt}")
```

## 4. Таблиця результатів: фрукт → кількість

```python
# Таблиця відсортована за кількістю (спадання)
sorted_results = sorted(counter.items(), key=lambda x: x[1], reverse=True)

print(f"{'Фрукт':<12} {'Кількість':>10} {'Візуалізація'}")
print("-" * 40)
for fruit, cnt in sorted_results:
    bar = "█" * cnt
    print(f"{fruit:<12} {cnt:>10}   {bar}")
```

## 5. Фрукт-лідер

```python
# Визначення лідера
leader = counter.most_common(1)[0]

print(f"🏆 Фрукт-лідер: '{leader[0]}' — {leader[1]} голосів")
print(f"   Це {leader[1] / len(votes) * 100:.1f}% від усіх голосів ({len(votes)} шт.)")
```

## 6. Топ-3 фрукти

```python
# Топ-3 через most_common()
top3 = counter.most_common(3)

medals = ["🥇", "🥈", "🥉"]
print("Топ-3 фрукти:")
print(f"{'Місце':<6} {'Фрукт':<12} {'Голосів':>8} {'%':>8}")
print("-" * 38)
for i, (fruit, cnt) in enumerate(top3):
    pct = cnt / len(votes) * 100
    print(f"{medals[i]:<7} {fruit:<12} {cnt:>7}   {pct:>6.1f}%")
```

* Для кожного предмета обчисліть:
   * середній бал;
   * мінімальний бал;
   * максимальний бал.

```python
math    = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]
```

## 1. Обчислення показників для кожного предмета

```python
subjects = {
    "Math": math,
    "Physics": physics,
    "English": english
}

stats = {}

for name in subjects:
    scores = subjects[name]

    avg = sum(scores) / len(scores)
    mn = min(scores)
    mx = max(scores)

    stats[name] = {
        "avg": avg,
        "min": mn,
        "max": mx
    }

print("Статистика по предметах:")
for name in stats:
    s = stats[name]
    print(name, "avg =", round(s["avg"], 1), "min =", s["min"], "max =", s["max"])

```

## 2. Предмет з найвищим середнім балом

```python
best_subject = max(stats, key=lambda s: stats[s]["avg"])
print(best_subject)
```

## 3. Таблиця результатів

```python
col_w = 10
header = f"{'Предмет':<12} {'Середній':>{col_w}} {'Мін':>{col_w}} {'Макс':>{col_w}} {'Рейтинг'}"


print("-------------------------------------------------------")
print(header)
print("-------------------------------------------------------")

sorted_subjects = sorted(stats.items(), key=lambda x: x[1]["avg"], reverse=True)
medals = ["🥇", "🥈", "🥉"]

for i, (name, s) in enumerate(sorted_subjects):
    bar = "█" * int(s["avg"] // 10)
    medal = medals[i] if i < 3 else "  "
    print(f"{name:<12} {s['avg']:>{col_w}.1f} {s['min']:>{col_w}} {s['max']:>{col_w}}   {medal} {bar}")

print("-------------------------------------------------------")
print(f"\nНайкращий предмет: {best_subject} — середній бал {stats[best_subject]['avg']:.1f}")
```

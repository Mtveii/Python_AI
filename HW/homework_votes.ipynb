```python
from collections import Counter

votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
```

## 1. Підрахунок через `count()`

```python
count_method = {}

for fruit in votes:
    if fruit in count_method:
        count_method[fruit] += 1
    else:
        count_method[fruit] = 1

print("Метод count:")
for fruit in count_method:
    print(fruit, ":", count_method[fruit])
```

## 2. Підрахунок через `dict`

```python
count_dict = {}

for fruit in votes:
    if fruit in count_dict:
        count_dict[fruit] += 1
    else:
        count_dict[fruit] = 1

for fruit in count_dict:
    print(fruit, ":", count_dict[fruit])
```

## 3. Підрахунок через `collections.Counter`

```python
count = {}

for fruit in votes:
    if fruit in count:
        count[fruit] += 1
    else:
        count[fruit] = 1

for fruit in count:
    print(fruit, ":", count[fruit])
```

## 4. Таблиця результатів: фрукт → кількість

```python
results = list(counter.items())
results.sort(reverse=True)

print("Фрукт   Кількість  ")
print("----------------------------------")

for item in results:
    fruit = item[0]
    cnt = item[1]

    print(fruit, cnt)
```

## 5. Фрукт-лідер

```python
leader = ""
max_count = 0

for fruit in counter:
    if counter[fruit] > max_count:
        max_count = counter[fruit]
        leader = fruit

print("Фрукт-лідер:", leader)
print("Голосів:", max_count)

percent = max_count / len(votes) * 100
print("Це", round(percent, 1), "% від усіх голосів")
```

## 6. Топ-3 фрукти

```python
print("Топ-3 фрукти:")

place = 1

for fruit in counter:
    if place > 3:
        break

    cnt = counter[fruit]
    percent = cnt / len(votes) * 100

    print(place, fruit, cnt, round(percent, 1), "%")

    place += 1
```

# 📊 Homework: Scores Comparison
Порівняння результатів з предметів та обчислення основних показників

```python
# Дані — оцінки по предметах
math    = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]
```

## 1. Обчислення показників для кожного предмета

```python
subjects = {
    "Math":    math,
    "Physics": physics,
    "English": english,
}

stats = {}
for name, scores in subjects.items():
    stats[name] = {
        "avg": sum(scores) / len(scores),
        "min": min(scores),
        "max": max(scores),
    }

print("Статистика по предметах:")
for name, s in stats.items():
    print(f"  {name}: avg={s['avg']:.1f}, min={s['min']}, max={s['max']}")
```

## 2. Предмет з найвищим середнім балом

```python
best_subject = max(stats, key=lambda s: stats[s]["avg"])
print(best_subject)
```

## 3. Таблиця результатів

```python
col_w = 10
header = f"{'Предмет':<12} {'Середній':>{col_w}} {'Мін':>{col_w}} {'Макс':>{col_w}} {'Рейтинг'}"
sep    = "-" * (len(header) + 2)

print(sep)
print(header)
print(sep)

sorted_subjects = sorted(stats.items(), key=lambda x: x[1]["avg"], reverse=True)
medals = ["🥇", "🥈", "🥉"]

for i, (name, s) in enumerate(sorted_subjects):
    bar = "█" * int(s["avg"] // 10)
    medal = medals[i] if i < 3 else "  "
    print(f"{name:<12} {s['avg']:>{col_w}.1f} {s['min']:>{col_w}} {s['max']:>{col_w}}   {medal} {bar}")

print(sep)
print(f"\n🏆 Найкращий предмет: {best_subject} — середній бал {stats[best_subject]['avg']:.1f}")
```

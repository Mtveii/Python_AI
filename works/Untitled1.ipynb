### Завдання 3
Створіть масив розміром 3×5, заповнений числами від 0 до 14.
* Виведіть вихідний масив.
* Виконайте транспонування масиву (рядки стають стовпцями, а стовпці — рядками).
* Застосуйте операцію reshape, щоб отримати масив 5×3.
* Відобразіть результати на екран.

```python
rand_3_5 = np.random.randint(0, 14, size=(3, 5))
rand_3_5
```

```python
print("Форма масиву", rand_3_5.shape)
```

```python
trans = np.transpose(rand_3_5)
print("Транспонування:\n", trans)
print("Форма транспоногоного масиву:\n", np.transpose(rand_3_5))
```

```python
resh = np.reshape(rand_3_5, shape=(5, 3))
print("Зміна розмірності (решейпінг):\n", resh)
print("Форма нового масиву:", resh.shape)
```
